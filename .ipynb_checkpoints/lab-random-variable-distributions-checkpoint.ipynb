```python
#1. Import the necessary libraries.
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cust_df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
cust_df.columns=pd.Series(cust_df.columns).apply(lambda x: "_".join(y.lower() for y in x.split(" ")))
cust_df.head()
```

```python
#1. Get the numerical variables from our dataset.
cust_df_num = cust_df.select_dtypes(include = np.number)
cust_df_num.head()
```

```python
#2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
for col in cust_df_num.columns:
    sns.displot(cust_df_num[col],kde=True)
    plt.show()
```

We observe that CLV, income, monthly_premium_auto, total_claim_amount - all have partial or full exponential distribution nature. We'll now apply the log transform and see if the distribution takes a shift towards normality.

The column customer_lifetime_value and total_claim_amount (although it's an output variable) fits closest to the exponential distribution. We'll apply the below transformations to see if these columns bring the distribution back to normality.

```python
def log_transform_clean(x):
    x = np.log10(x)
    if np.isfinite(x):
        return x
    else:
        return np.NAN

def quantile_transform_clean(df,col_name):
    iqr = np.percentile(df[col_name],75) - np.percentile(df[col_name],25)
    upper_limit = np.percentile(df[col_name],75) + 1.5*iqr
    lower_limit = np.percentile(df[col_name],25) - 1.5*iqr
    return df[(df[col_name]>lower_limit) & (df[col_name]<upper_limit)]

def boxcox_transform_clean(df,col_name):
    df[col_name] = np.where(df[col_name]<=0,0,df[col_name])
    return stats.boxcox(df[col_name])[0]
```

```python
for col in ['customer_lifetime_value','total_claim_amount']:
    # Log Transform
    print("Log Transform:")
    cust_df_num[col+'_log'] = list(map(log_transform_clean, cust_df_num[col]))
    cust_df_num[col+'_log'] = cust_df_num[col+'_log'].fillna(np.mean(cust_df_num[col]))
    sns.displot(cust_df_num[col+'_log'])
    plt.show()
    
    # Quantile Transform
    print("Quantile Transform:")
    cust_df_num_quant = quantile_transform_clean(cust_df_num,col)
    sns.displot(cust_df_num_quant[col])
    plt.show()
    
    # BoxCox Transform
    print("BoxCox Transform:")
    cust_df_num_boxcox = boxcox_transform_clean(cust_df_num,col)
    sns.displot(cust_df_num_boxcox)
    plt.show()
```

As observed above, the boxcox transform performs the best in shifting the distribution towards normality, followed by log transform and then quantile tranform.
